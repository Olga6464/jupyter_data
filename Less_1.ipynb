#### Парсинг API

1. Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.


```python
import requests
import json
from pprint import pprint
```

```python
main_link = 'http://api.github.com/users'
params = 'Olga6464'
```

```python
response = requests.get(main_link + '/' + params +'/repos')
```

```python
j_data = response.json()
```

```python
pprint(j_data )
```

```python
pprint (f'Список репозиториев {params}: ')
for i in j_data:
    pprint(i['name'])
```

```python
with open('j_data.json', 'w') as f:
    json.dump(j_data, f)
```
